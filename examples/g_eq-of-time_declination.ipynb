```python
import numpy as np
import matplotlib.pyplot as plt
import solarpy.radiation as spr
```

# G - extraterrestrial radiation on normal plane

Intensity (usually expressed in irradiance units W/m2) of the Sun at the top of the Earth’s atmosphere, on a plane normal to the solar vector. Also called [solar irradiance](https://en.wikipedia.org/wiki/Solar_irradiance). It changes throughout the year as the Earth-Sun distance varies due to the Earth’s elliptical orbit.

```python
# create the time-series
n = range(1, 366)
Gon_ = []

for i in n:
    Gon_.append(spr.Gon(i))

# plot the series
plt.figure(figsize=(18, 6))
plt.plot(n, Gon_, label='Gon')
plt.xlim(0, 366), plt.ylim(1320, 1420)
plt.xlabel('day of the year', fontsize=15)
plt.ylabel('W/m2', fontsize=15)
plt.title('Extraterrestrial radiation on normal plane', fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```

# E - equation of time

Due to the eccentricity of the Earth's orbit "around" the Sun and the obliquity of the ecliptic, there is a [difference](https://aa.usno.navy.mil/faq/docs/eqtime.php) between _apparent solar time_ and _mean solar time_. This difference changes, and it is defined by the **equation of time**:

```python
# create the time-series
n = range(1, 366)
E_ = []

for i in n:
    E_.append(spr.Eq_time(i))

# plot the series
plt.figure(figsize=(18, 6))
plt.plot(n, E_, label='E')
plt.xlim(0, 366), plt.ylim(-15, 17)
plt.xlabel('day of the year', fontsize=15)
plt.ylabel('min', fontsize=15)
plt.title('Equation of time', fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```

# Declination

Due to the obliquity of the ecliptic (Earth's equator plane 23.45º tilting wrt the plane of the Earth's orbit around the Sun), the angle between the solar beam and the plane of the Earth's equator changes. This angle is called **declination**, and it varies from 23.45 to -23.45 degrees. 

```python
# create the time-series
n = range(1, 366)
delta_ = []

for i in n:
    delta_.append(np.rad2deg(spr.declination(i)))

# plot the time-series
plt.figure(figsize=(18, 6))
plt.plot(n, delta_, label='delta')
plt.xlim(0, 366), plt.ylim(-25, 25)
plt.xlabel('day of the year', fontsize=15)
plt.ylabel('degrees', fontsize=15)
plt.title('Sun declination', fontsize=18)
plt.legend()
plt.grid(True)
plt.show()
```
